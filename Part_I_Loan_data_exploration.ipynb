## Univariate Exploration


First, let's look at the distribution of Loan status.

```python
plt.figure(figsize=[20, 5])
base_color = sb.color_palette()[0]

plt.subplot(1,2,1)
g = sb.countplot(y = 'LoanStatus', data = loan_dfc, color = base_color)
g.set(title = 'Distribution of Loan Status') ;

plt.subplot(1,2,2)
z = loan_dfc.loc[(loan_dfc.LoanStatus == 'Chargedoff') | (loan_dfc.LoanStatus == 'Completed') | (loan_dfc.LoanStatus == 'Current') | (loan_dfc.LoanStatus == 'Defaulted')]
g = sb.countplot(y = 'LoanStatus', data = z, color = base_color)
g.set(title = 'Distribution of Loan Status');
```

We can see the major categories are 'Completed', 'Current', 'Defaulted' and 'Chargedoff'.

```python
plot_o = ['$100,000+', '$75,000-99,999', '$50,000-74,999', '$25,000-49,999', '$1-24,999', '$0', 'Not employed']
g = sb.countplot(y = 'IncomeRange', data = loan_dfc, color = base_color, order=plot_o)
g.set(title = 'Distribution of Income Range');
```

We can see that the 25,000 - 49,999 and the 50,000 - 74,999 are the 2 major income categories.

```python
g = sb.countplot(y = 'EmploymentStatus', data = loan_dfc, color = base_color)
g.set(title = 'Distribution of Employment Status');
```

We can see that majority are employed.

```python
plt.figure(figsize=[20, 5])

plt.subplot(1,2,1)
g = sb.countplot(x = 'IsBorrowerHomeowner', data = loan_dfc, color = base_color)
g.set(title = 'Distribution of Home owners');

plt.subplot(1,2,2)
g = sb.countplot(x = 'CurrentlyInGroup', data = loan_dfc, color = base_color)
g.set(title = 'Distribution of People in a Group');
```

We can see that there are lightly more home owners however, most borrowers are not in a group.

```python
plt.figure(figsize=[15,5])

plt.subplot(1,2,1)
plt.hist(x = 'LoanOriginalAmount', data = loan_dfc)
plt.xlabel('LoanOriginalAmount')
plt.title('Distribution of Loan Amount');

plt.subplot(1,2,2)
g = sb.distplot(loan_dfc['LoanOriginalAmount'])
g.set(title='Distribution of Loan Amount');
```

We can see that most loan requests are for 4,000 dollars, followed by 15,000 and 10,000 in that order.

```python
loan_O = loan_dfc.groupby('Occupation').size()
loan_O = loan_O.reset_index(name = 'counts')
df_o = loan_O.nlargest(10, 'counts')

df_o
```

## Bivariate Exploration

> 

```python
loan_dfc.head()
```

```python
loandf_counts = z.groupby(['LoanStatus','IncomeRange']).size()
loandf_counts = loandf_counts.reset_index(name ='counts')
loandf_counts = loandf_counts.pivot(index = 'IncomeRange', columns= 'LoanStatus', values='counts')

loandf_counts
```

```python
prop = loandf_counts.copy()
prop['Totals'] = prop.sum(axis=1)

prop['Chargedoff'] = prop['Chargedoff']/prop['Totals']
prop['Completed'] = prop['Completed']/prop['Totals']
prop['Current'] = prop['Current']/prop['Totals']
prop['Defaulted'] = prop['Defaulted']/prop['Totals']
prop.drop('Totals', axis = 1, inplace=True)
prop
```

```python
plt.figure(figsize=[12,10])

plt.subplot(2,1,1)
g = sb.heatmap(loandf_counts, annot=True)
g.set(title = 'Income Range Vs Loans Status')

plt.subplot(2,1,2)
h = sb.heatmap(prop, annot=True)
h.set(title = 'Income Range Vs Loans Status Proportions');
```

The heat map shows a trend of high income range having a percentage of borrowers with current loans.

```python
e_status = z.groupby(['LoanStatus','EmploymentStatus']).size()
e_status = e_status.reset_index(name ='counts')
e_status = e_status.pivot(index = 'EmploymentStatus', columns= 'LoanStatus', values='counts')

e_status
```

```python
prop_e = e_status.copy()
prop_e['Totals'] = prop_e.sum(axis=1)

prop_e['Chargedoff'] = prop_e['Chargedoff']/prop_e['Totals']
prop_e['Completed'] = prop_e['Completed']/prop_e['Totals']
prop_e['Current'] = prop_e['Current']/prop_e['Totals']
prop_e['Defaulted'] = prop_e['Defaulted']/prop_e['Totals']
prop_e.drop('Totals', axis = 1, inplace=True)
prop_e
```

```python
plt.figure(figsize=[12,10])

plt.subplot(2,1,1)
g = sb.heatmap(e_status, annot=True)
g.set(title = 'Employment Status Vs Loans Status')

plt.subplot(2,1,2)
h = sb.heatmap(prop_e, annot=True)
h.set(title = 'Employment Status Vs Loans Status Proportions');
```

This shows that loans are mostly given to borrowers who are employed.

```python
g = sb.countplot(data = z, x = 'LoanStatus', hue = 'IsBorrowerHomeowner')
g.set(title = 'Comparing Loan Status of Home Owners Vs Non Home Owners');
```

This shows that with the current loans, more home owners were given loan compared to none home owners.

```python
g = sb.boxplot(data = z, x = 'LoanStatus', y = 'LoanOriginalAmount')
g.set(title ='Distribution of Loan Status With respect to Loan Amount in $');
```

This shows that most of the loans completed were about 15,000 dollars and below however, when compared with the current loans given, about 75% are 15,000 dollar or lower.

## Multivariate Exploration

```python
g = sb.relplot(x = 'LoanOriginalAmount', y = 'BorrowerRate', data = z, hue = 'IsBorrowerHomeowner')
g.set(title = 'Relationship between Loan Amount, Interest Rate and Home Owner Statues');
```

This shows that rates on loans above 25,000 dollars fall below 0.2 and are mostly assessed by home owners.

```python
g = sb.FacetGrid(data = z, col='LoanStatus')
g.map(sb.violinplot, 'LoanOriginalAmount', 'EmploymentStatus');
```

While this shows that the 'Not empolyed' status has the highest distribution on completed loans, it also shows that 'Not employed' status has the highest distribution of deafalts.
Looking at the current loans, full-time status has the highest distribution.

```python
g = sb.FacetGrid(data = z, col='LoanStatus')
g.map(sb.violinplot, 'LoanOriginalAmount', 'IncomeRange');
```

Looking at the income range, the 25,000 - 49,999 dollars range have the highest distribution on completed loan payments.


```python
g = sb.relplot(x = 'LoanOriginalAmount', y = 'CreditScoreRangeUpper', data = z, hue = 'LoanStatus')
g.set(title = 'Relationship between Loan Amount, Credit Score Range and Loan Statues');
```

We see here that loans above 25,000 dollars are only assesed by credit scores above 740 and are mostly current loans.

## Conclusions
>It can be observed from these visualizations that current loans and loans above 25,000 dollars are given to borrowers with the most favorable characteristics to complete the loans like, higher income range, higher credit scores and home owners.
